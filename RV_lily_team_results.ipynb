```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import tqdm as tqdm
from tqdm import tqdm
import pandas as pd
from datetime import datetime
from scipy.interpolate import interp1d

plt.rcParams['font.size'] = 16
plt.rc("font", family="serif")

from os import walk
```

```python
files = [
    "lily_stellar_activity_data/10700_activity.csv",    # 0. 
    "lily_stellar_activity_data/26965_activity.csv",    # 1.
    "lily_stellar_activity_data/34411_activity.csv",    # 2. First one used
    "lily_stellar_activity_data/101501_activity.csv"    # 3.
]

def plot_stuff1(df):
    plt.figure(figsize=(25,6))
    rv = df["CBC RV [m/s]"]
    rv_err = df["CBC RV Err. [m/s]"]
    t = df["Time [MJD]"]
    t = t - min(t)
    plt.errorbar(t, rv, yerr=rv_err, fmt=".", label=f"Mean error = {np.mean(rv_err):.3}")
    plt.xlabel("Time [MJD]")
    plt.ylabel("CBC RV [m/s]")
    plt.legend()


def plot_stuff2(df):
    plt.figure(figsize=(25,6))
    rv = df["CCF RV [m/s]"]
    rv_err = df["CCF RV Err. [m/s]"]
    t = df["Time [MJD]"]
    plt.errorbar(t, rv, yerr=rv_err, fmt=".")
    plt.xlabel("Time [MJD]")
    plt.ylabel("CBC RV [m/s]")


# plot_stuff1(pd.read_csv(files[0]))
# plot_stuff1(pd.read_csv(files[1]))
plot_stuff1(pd.read_csv(files[2]))
# plot_stuff1(pd.read_csv(files[3]))
```
